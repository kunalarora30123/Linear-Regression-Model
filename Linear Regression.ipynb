# Linear Regression 

```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv("boston.csv")
df
```

```python
df.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
df
```

```python
# Independent and Dependent Data
x= df.drop("Price",axis=1)
y= df.Price
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
x_scaled=StandardScaler().fit_transform(x)
```

```python
for i in df.columns:
    if i!="Price":
        df[i]=StandardScaler().fit_transform(df[[i]])
        df=df[df[i].between(-3,3)]
```

```python
df
```

```python
simple_x= x[["CRIM"]] # Mx+c
# Multiple Regression [M1*x1 + M2*x2 +M3*x3+.....+C]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0,test_size=0.2)
```

```python
(y_test.shape[0]/(y_test.shape[0]+y_train.shape[0]))*100
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR= LinearRegression()
```

```python
LR
```

```python
LR.fit(x_train,y_train)
```

```python
predictions=LR.predict(x_test)
```

```python
predictions
```

```python
pd.DataFrame({"Actual":y_test,"Prediction":predictions})
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test,predictions)
```

```python
mean_absolute_error(y_test,predictions)
```

```python
import math
```

```python
math.sqrt(mean_squared_error(y_test,predictions))
```

```python
r2_score(y_test,predictions)
```

```python
LR.score(x_train,y_train)
```
